```python
from sklearn.metrics.pairwise import cosine_similarity


def find_best_chunks_cos_dist(
    themes, theme_embeddings, chunk_embeddings, chunk_texts, threshold=0.7
):
    for index in range(len(themes)):
        # Compute cosine similarity between the current theme and all chunks
        cosine_similarities = cosine_similarity(
            theme_embeddings[index].reshape(1, -1), chunk_embeddings
        ).flatten()

        # Combine chunk texts and their corresponding cosine similarity scores
        chunk_similarity_pairs = list(zip(chunk_texts, cosine_similarities))

        # Filter pairs to keep only those with scores above the threshold
        filtered_chunk_similarity_pairs = [
            (chunk, score)
            for chunk, score in chunk_similarity_pairs
            if score > threshold
        ]

        # Sort the pairs based on cosine similarity scores in descending order
        sorted_chunk_similarity_pairs = sorted(
            filtered_chunk_similarity_pairs, key=lambda x: x[1], reverse=True
        )

        # Display the best scoring chunk and its score if any
        if sorted_chunk_similarity_pairs:
            print(f"Theme: {themes[index]}")
            best_chunk, best_score = sorted_chunk_similarity_pairs[0]
            print(f"Best Score: {best_score:.4f}, Best Chunk: {best_chunk}\n")


find_best_chunks_cos_dist(themes, theme_embeddings, chunk_embeddings, chunk_texts)
```
